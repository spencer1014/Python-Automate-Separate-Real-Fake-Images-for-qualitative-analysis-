```python
import os
import shutil # high level operation on a file like a copy, create, and remote
import pandas as pd
```

```python
csv_file_path = r"C:\Users\User\Desktop\PriorityDynamics\cidaut ai fake scene\data_overall.csv"
images_folder_path = r"C:\Users\User\Desktop\PriorityDynamics\cidaut ai fake scene\images"
real_folder_path = r"C:\Users\User\Desktop\PriorityDynamics\cidaut ai fake scene\real"
fake_folder_path = r"C:\Users\User\Desktop\PriorityDynamics\cidaut ai fake scene\fake"
```

```python
df = pd.read_csv(csv_file_path)
```

```python
os.makedirs(fake_folder_path, exist_ok=True)
os.makedirs(real_folder_path, exist_ok=True)
```

```python
for _, row in df.iterrows():
    image_name = row['image']
    image_class = row['label']

    image_path = os.path.join(images_folder_path, image_name)

    print(image_path)
    
    if image_class == 0 :
        shutil.move(image_path, os.path.join(fake_folder_path, image_name))

    elif image_class == 1 :
        shutil.move(image_path, os.path.join(real_folder_path, image_name))

print("Image have been successfully moved")
```
