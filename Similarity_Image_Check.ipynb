```python
import pandas as pd
import shutil
import os
import cv2
from skimage.metrics import structural_similarity as ssim
import numpy as np
from pathlib import Path
```

```python
images_folder_path = r"D:\PriorityDynamics\cidaut ai fake scene\python\Images"
images_copy_folder = r"D:\PriorityDynamics\cidaut ai fake scene\python\Images Copy"
folder_path = Path(images_copy_folder)
separated_folder = r"D:\PriorityDynamics\cidaut ai fake scene\python\Images separated"
```

```python
combined = os.listdir(images_folder_path)
```

```python
print(len(combined))
```

```python
Image_1 = []
Image_2 = []
similar = []
```

```python
for i in combined:
    for e in combined:

        if i == e:
            continue
        
        image_a_path1 = os.path.join(images_folder_path, i)
        image_b_path2 = os.path.join(images_folder_path, e)

        image_a = cv2.imread(image_a_path1)
        image_b = cv2.imread(image_b_path2)

        image_a1 = cv2.cvtColor(image_a, cv2. COLOR_BGR2GRAY)
        image_b2 = cv2.cvtColor(image_b, cv2. COLOR_BGR2GRAY)

        similarity, _ = ssim(image_a1, image_b2, full=True)

        if similarity > 0.9:
            print(similarity)
            print(i)
            print(e)
            Image_1.append(i)
            Image_2.append(e)
            similar.append(similarity)
```

```python
data = {'Image 1': Image_1, 'Image 2': Image_2, 'Similarity': similar}
```

```python
df = pd.DataFrame(data)
```

```python
if folder_path.is_dir(): 
    print("Folder is existo")
else:
    shutil.copytree(images_folder_path, images_copy_folder)
    Print("Copying")
    
```

```python
for index, row in df.iterrows():
    folder_name = f"Pair_{row['Image 1']}_&_{row['Image 2']}"
    target_path = os.path.join(separated_folder, folder_name)
    os.makedirs(target_path, exist_ok=True)

    for col in ['Image 1', 'Image 2']:
        img_name = row[col]
        src = os.path.join(images_copy_folder, img_name)
        dst = os.path.join(target_path, img_name)

        if os.path.exists(src):
            shutil.copy(src, dst)
        else:
            print(f"{img_name} not found")
```
